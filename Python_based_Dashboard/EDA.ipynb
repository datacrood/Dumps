```python
import plotly.graph_objects as go
import plotly.express as px

# Create a heatmap
def create_heatmap(df, selected_state, selected_district, selected_commodity):
    filtered_df = df[(df['census_state_name'] == selected_state) & 
                     (df['census_district_name'] == selected_district) & 
                     (df['commodity_name'] == selected_commodity)]
    
    pivot = filtered_df.pivot_table(values='modal_price', index='census_district_name', 
                                    columns=filtered_df['date'].dt.strftime('%B'), 
                                    aggfunc='mean')
    
    fig = px.imshow(pivot, labels=dict(color="Average Price"),
                    title=f"Monthly Average Modal Price for {selected_commodity} in {selected_state}")
    return fig

# Create a line chart
def create_line_chart(df, selected_state, selected_district, selected_commodity):
    filtered_df = df[(df['census_state_name'] == selected_state) & 
                     (df['census_district_name'] == selected_district) & 
                     (df['commodity_name'] == selected_commodity)]
    
    monthly_avg = filtered_df.groupby(filtered_df['date'].dt.strftime('%B'))['modal_price'].mean().reset_index()
    
    fig = px.line(monthly_avg, x='date', y='modal_price', 
                  title=f"Monthly Average Modal Price Trend for {selected_commodity} in {selected_state}")
    return fig
```
